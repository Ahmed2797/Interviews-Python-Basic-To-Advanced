    Problem Name: The Minion Game
    Source: HackerRank
    Key Concept:
    - Substring counting
    - Vowel vs Consonant
    - Math optimization (no substring generation)
    - Time Complexity: O(n)


    ‚Ä¢ Substring contribution technique
    ‚Ä¢ Each index contributes (len(string) - index)
    ‚Ä¢ No substring generation
    ‚Ä¢ Time Complexity: O(n)
    ‚Ä¢ Space Complexity: O(1)


```python
def minion_game(string):
    vowels = "AEIOU"
    kevin = 0
    stuart = 0
    n = len(string)

    for i in range(n):
        if string[i] in vowels:
            kevin += n - i
        else:
            stuart += n - i

    if kevin > stuart:
        print("Kevin", kevin)
    elif stuart > kevin:
        print("Stuart", stuart)
    else:
        print("Draw")


if __name__ == '__main__':
    s = "BANANA"
    minion_game(s)
```

    üìù Question: Default Arguments (Even & Odd Stream)

    You are given two classes, EvenStream and OddStream, which generate even and odd numbers respectively.

    Class definitions:

    - EvenStream

    Starts from 0

    Each call to `get_next()` returns the next even number

    - OddStream

    Starts from 1

    Each call to `get_next()` returns the next odd number
        
    Problem: Print Even/Odd Stream
    Key Concept:
    - Default argument pitfall in Python
    - Use None instead of object in function parameter
    - State isolation using new instances


```python
class EvenStream:
    def __init__(self):
        self.current = 0

    def get_next(self):
        value = self.current
        self.current += 2
        return value


class OddStream:
    def __init__(self):
        self.current = 1

    def get_next(self):
        value = self.current
        self.current += 2
        return value


def print_from_stream(n, stream=None):
    if stream is None:
        stream = EvenStream()

    for _ in range(n):
        print(stream.get_next())


# üîΩ THIS PART WAS MISSING üîΩ
if __name__ == "__main__":
    q = int(input().strip()) ## 2 how many number print

    for _ in range(q):
        parts = input().split() ##  odd 2  ## even 2

        stream_type = parts[0]
        n = int(parts[1])

        if stream_type == "odd":
            print_from_stream(n, OddStream())
        else:
            print_from_stream(n)
```

```python
def is_number_palindrome(n):
    return str(n) == str(n)[::-1]

# Test
print(is_number_palindrome(121))    # True
print(is_number_palindrome(12321))  # True
print(is_number_palindrome(123))    # False


def is_palindrome(word):
    word = word.lower()  
    left = 0
    right = len(word) - 1

    while left < right:
        if word[left] != word[right]:
            return False
        left += 1
        right -= 1

    return True

# Test
print(is_palindrome("‡¶ï‡¶æ‡¶ï"))      # True
print(is_palindrome("‡¶®‡¶Æ‡¶®"))     # True
print(is_palindrome("banana"))  # False
```
