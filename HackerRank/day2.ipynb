```python
if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        print(name,line)
        scores = list(map(float, line))
        print(scores)
        student_marks[name] = scores
    query_name = input()
    print(query_name)
    result = [(name,score) for name,score in student_marks.items() if name == query_name]
    print(result[0])
    marks = [score for score in result[0][1]]
    print(marks)
    avg = sum(marks) / len(marks)
    print(avg)
```

```python
num = []

if __name__ == '__main__':
    for _ in range(int(input())):
        N = input().split()
        # print(N)
        if N[0] == "append":
            num.append(int(N[1]))
        elif N[0] == "insert":
            num.insert(int(N[1]),int(N[2]))
        elif N[0] == "remove":
            num.remove(int(N[1]))
        elif N[0] == "pop":
            num.pop()
        elif N[0] == "sort":
            num.sort()
        elif N[0] == "reverse":
            num.reverse()
        elif N[0] == "print":
            print(num)
        else:
            print("Enter Input....")


print(num)
```
