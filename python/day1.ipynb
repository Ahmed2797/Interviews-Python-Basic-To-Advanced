### Mutable
In Python, mutable objects can be changed after creation. Examples include lists, dictionaries, and sets. For instance, you can modify a list's elements without creating a new list.
### Imutable
In contrast, immutable objects cannot be altered once created, such as strings, tuples, and integers. If you try to change them, a new object is made.



```python
# Mutable example: list
my_list = [1, 2, 3]
my_list[0] = 10  # Changes the original list
print(my_list)  # Output: [10, 2, 3]

# Immutable example: string
my_str = "hello"
my_str = my_str.upper()  # Creates a new string
print(my_str)  # Output: "HELLO"
```

```python
# Simple Example to Understand Frequency Count
# Let's use a smaller list: [1, 2, 2, 3]
# Step by step:

small_list = [1, 2, 2, 3]
freq_simple = {}

print("Starting with empty dictionary: freq_simple =", freq_simple)

for i,num in enumerate(small_list):
    print(f"\nProcessing number: {i,num}")
    current_count = freq_simple.get(num, 0)  # Get current count (0 if not found)
    print(f"Current count for {num}: {current_count}")
    
    freq_simple[num] = current_count + 1
    print(f"Updated freq_simple: {freq_simple}")

print(f"\nFinal result: {freq_simple}")
# Output: {1: 1, 2: 2, 3: 1}
```

```python
fruits = ['apple', 'banana', 'apple', 'cherry']
freq = {}

for i,fruit in enumerate(fruits):
    print(f"\nProcessing number: {i,fruit}")
    current_count = freq.get(fruit, 0)
    print(f"Current count for {fruit}: {current_count}")
    
    freq[fruit] = current_count + 1
    print(f"Updated freq_simple: {freq}")

print(freq)  # Output: {'apple': 2, 'banana': 1, 'cherry': 1}
```

### list comprehension

```python
a = [1,2,3,4,5,6]
result = [i*2 for i in range(len(a))]
print(result)

res = [i*2 for i in a]
print(res)
```

```python
# Easy Explanation of List Comprehension
# List comprehension is a short way to create a new list from an existing list or range.

# Basic syntax: [expression for item in iterable]

# Example 1: Double each number in a list
numbers = [1, 2, 3, 4, 5]
doubled = [x * 2 for x in numbers]
print("Doubled:", doubled)  # [2, 4, 6, 8, 10]

# Example 2: Square each number
squared = [x ** 2 for x in numbers]
print("Squared:", squared)  # [1, 4, 9, 16, 25]

# Example 3: Get even numbers only
evens = [x for x in numbers if x % 2 == 0]
print("Evens:", evens)  # [2, 4]

# Your code: [i*2 for i in range(len(a))]
# range(len(a)) gives [0,1,2,3,4,5] for a=[1,2,3,4,5,6]
# So i*2 for i in [0,1,2,3,4,5] = [0,2,4,6,8,10]

# Better way: [x*2 for x in a] to double the list elements
better_result = [x*2 for x in a]
print("Better result:", better_result)  # [2,4,6,8,10,12]
```

```python
import re
from collections import Counter

def text_analyzer(text: str) -> dict:
    # Edge case: empty or invalid input
    if not isinstance(text, str) or not text.strip():
        return {
            "total_words": 0,
            "unique_words": 0,
            "top_5_words": [],
            "average_word_length": 0.0
        }

    # Normalize text: lowercase + remove punctuation/special chars
    words = re.findall(r"[a-zA-Z]+", text.lower())

    # Edge case: no valid words after cleaning
    if not words:
        return {
            "total_words": 0,
            "unique_words": 0,
            "top_5_words": [],
            "average_word_length": 0.0
        }

    total_words = len(words)
    unique_words = len(set(words))

    counts = Counter(words)
    top_5_words = counts.most_common(5)

    avg_word_length = round(
        sum(len(w) for w in words) / total_words, 2
    )

    return {
        "total_words": total_words,
        "unique_words": unique_words,
        "top_5_words": top_5_words,
        "average_word_length": avg_word_length
    }
text = "Hello, hello! Python is great. Python is FUN!!!"
# print(text_analyzer(text))
```
