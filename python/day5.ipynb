‚ùì What is Closure?

üëâ `Inner function remembers outer function variables even after execution.`

```python
def power(n):
    def inner(x):
        return x ** n
    return inner

square = power(2)
cube = power(3)

print(square(5))  # 25
print(cube(5))    # 125

```

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Time: {end - start}")
        return result
    return wrapper

@timer
def slow():
    time.sleep(2)

slow()
```

```python
# Create a decorator that:
# - counts how many times a function is called
# - stores count inside closure


def multi(a: int):
    count = 0

    def add(b: int):
        nonlocal count
        count += 1
        return a + b, count

    return add

adder = multi(10)

print(adder(5))   # (15, 1)
print(adder(10))  # (20, 2)
print(adder(20))  # (30, 3)


## using class methold
class Multi:
    def __init__(self, a):
        self.a = a
        self.count = 0

    def add(self, b):
        self.count += 1
        return self.a + b, self.count


```

```python
# Create a decorator that:
# - counts function calls
# - does NOT use global variables

def ahmed(*args, **kwargs):
    count = 0

    def tanvir(*vals):
        nonlocal count
        count += 1
        return count, vals, args, kwargs

    return tanvir

f = ahmed(1, 2, a=10)

print(f(100))
print(f(200, 300))
print(f("x"))
```

`add = call_counter(add)`

üëâ ‡¶è‡¶ñ‡¶® add ‡¶Ü‡¶∞ ‡¶Ü‡¶ó‡ßá‡¶∞ add ‡¶®‡¶æ

üëâ add ‡¶è‡¶ñ‡¶® wrapper

```python
def call_counter(func):
    count = 0
    def wrapper(*args, **kwargs):
        nonlocal count
        count += 1
        print(f"Call {count} of {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

@call_counter
def add(a, b):
    return a + b

print(add(2, 3))
print(add(4, 5))


## without decorator
class Counter:
    def __init__(self):
        self.count = 0

    def __call__(self):
        self.count += 1
        return self.count

c = Counter()
print(c())
print(c())


class Multi:
    def __init__(self, a):
        self.a = a
        self.count = 0

    def add(self, b):
        self.count += 1
        return self.a + b, self.count


```
