## Static & Class Methods

üîπ Static Method

Object ‡¶õ‡¶æ‡ßú‡¶æ‡¶á call ‡¶ï‡¶∞‡¶æ ‡¶Ø‡¶æ‡ßü

```python
class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b

print(MathUtils.add(2, 3))
```

```python
class Employee:
    company = "Google"

    @classmethod
    def show_company(cls):
        return print(cls.company)
    
    def add(self):
        return self.company + "Tech"

print(Employee.show_company())
# em = Employee()
# print(em.add())
```

üîπ ‡¶ï‡ßÄ ‡¶π‡¶ö‡ßç‡¶õ‡ßá ‡¶è‡¶ñ‡¶æ‡¶®‡ßá?

- Employee.show_company()

‚Üí method ‡¶è‡¶∞ ‡¶≠‡ßá‡¶§‡¶∞‡ßá print(cls.company) ‡¶ö‡¶≤‡ßá

‚Üí ‡¶§‡¶æ‡¶á Google ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶π‡ßü

- ‡¶ï‡¶ø‡¶®‡ßç‡¶§‡ßÅ show_company() method ‡¶ï‡¶ø‡¶õ‡ßÅ return ‡¶ï‡¶∞‡ßá ‡¶®‡¶æ

‚Üí Python ‡¶è default return ‡¶π‡ßü None

`Method ‡¶è print ‡¶Ü‡¶õ‡ßá ‡¶ï‡¶ø‡¶®‡ßç‡¶§‡ßÅ return ‡¶®‡ßá‡¶á, ‡¶§‡¶æ‡¶á method call ‡¶ï‡¶∞‡¶≤‡ßá ‡¶™‡ßç‡¶∞‡¶•‡¶Æ‡ßá print ‡¶π‡ßü, ‡¶§‡¶æ‡¶∞‡¶™‡¶∞ None return ‡¶π‡ßü‡•§`

```python
class Employee:
    company = "Google"

    @classmethod
    def show_company(cls):
        return cls.company
    
    def add(self):
        return self.company + "Tech"

print(Employee.show_company())
```

```python
class Person:
    def __init__(self,name="Ahmed"):
        self.name = name
        self.__marks = 30
    
    def print_name(self):
        rt = self.name
        mark = self.__marks
        print(mark)
        print(rt)

class Student(Person):
    def __init__(self,name, marks):
        super().__init__(name)
        self.__marks = marks   # private variable
        self.name = name

    def show(self):
        print(self.name, self.__marks)

student = Student(marks=20,name='Tanvir')
student.show()
student.print_name()
```

üü¢ Problem 1: Design a Bank Account (Amazon ‚Äì Very Common)
‚ùì Problem

‡¶è‡¶ï‡¶ü‡¶æ BankAccount class ‡¶°‡¶ø‡¶ú‡¶æ‡¶á‡¶® ‡¶ï‡¶∞‡ßã ‡¶Ø‡ßá‡¶ñ‡¶æ‡¶®‡ßá:

account_holder

private balance

deposit()

withdraw()

get_balance()

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        return print('Added Money')

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            return print("Money withdraw")
        else:
            return print("Insufficient balance")

    def get_balance(self):
        return self.__balance
```

```python
account = BankAccount("Ahmed",1000)
deposit = account.deposit(500)
# print(deposit)

withdraw = account.withdraw(200)
withdraw
```

```python
account.get_balance()
```

üü¢ Problem 2: Employee Management System (Google)

‚ùì Problem

Employee (name, id)

Manager inherit ‡¶ï‡¶∞‡¶¨‡ßá

salary calculate method override ‡¶ï‡¶∞‡¶¨‡ßá

```python
class Employee:
    def __init__(self,id,salary):
        self.id = id
        self.salary = salary
    
    def details(self):
        return self.id,self.salary
    
    def get_salary(self):
        return 30000
    
    def get_salary_month(self):
        return 50000
    
    

class Manger(Employee):

    def get_salary(self):
        return 60000
    
    def get_salary_month(self):
        return super().get_salary_month()
```

```python
manager = Manger(id=1,salary=1000)
print("get_salary:",manager.get_salary())

print("get_salary_month:",manager.get_salary_month())
```

üü¢ Problem 3: Shape Area Calculator (Google Onsite)

‚ùì Problem

‡¶è‡¶ï‡¶ü‡¶æ base Shape class

Child class: Circle, Rectangle

area() method override ‡¶ï‡¶∞‡ßã

```python
class Shape:
    r = 10
    def area(self):
        return self.r**2


class Circle(Shape):
    def __init__(self):
        super().__init__()
    def area(self):
        return super().area()
```

```python
circle = Circle()
circle.area()
```

```python
class Shape:

    def area(self):
        pass

class Circle(Shape):
    def __init__(self):
        self.r = 5
    def area(self):
        return 3.14 * self.r**2
    

class Rectangle(Shape):
    def __init__(self,w,h):
        self.h = h
        self.w = w

    def area(self):
        return self.h * self.w
```

```python
rectangle = Rectangle(5,3)
print("rectangle:",rectangle.area())

circle = Circle()
print("circle:",circle.area())
```

üü¢ Problem 7: Library Management (Amazon)
‚ùì Problem

Book class

Library class

add_book(), show_books()

```python
class Book:
    def __init__(self, title):
        self.title = title

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def show_books(self):
        for b in self.books:
            print(b.title)

```

```python
book = Book("Machine Learning")

library = Library()
library.add_book(book)

library.show_books()
```

üü¢ Problem 8: Payment System (Google ‚Äì Design Thinking)

Payment base class

```python
class GooglePay:
    def pay(self,amount):
        pass

class CredictCard(GooglePay):
    def pay(self,amount):
        return f"Paid {amount} via Credit Card"

class UPI(GooglePay):
    def pay(self,amount):
        return f"Paid {amount} via UPI"
```
