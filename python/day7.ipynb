üìÖ Day 24 ‚Äî Decorators

üìå Notice

    Function can exist inside another function,
    inside another function excute if its call inside function

```python
def outer():
    print("Outer function")

    def inner():
        print("Inner function")

    inner() ## call or not call see the output

outer()
```

```python
def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

@my_decorator
def hello():
    print("Hello!")

hello()
```

```python
def my_decorator(func):
    def wrapper(name):
        print("Start")
        func(name)
        print("End")
    return wrapper

@my_decorator
def greet(name):
    print("Hello", name)

greet("Efrot")
```

```python
def decorator(func):
    def wrapper(*args, **kwargs):
        print("Start")
        func(*args, **kwargs)
        print("End")
        return func(*args, **kwargs)
    return wrapper

@decorator
def greet(name):
    print("Hello", name)

greet("Efrot")
```

üìÖ Day 25 ‚Äî async def vs def (Very Common Confusion)

    üîπ def (Synchronous) ‚û°Ô∏è Blocks execution
    üîπ async def (Asynchronous) ‚û°Ô∏è Non-blocking

Async is NOT multithreading

| Async         | Thread           |
| ------------- | ---------------- |
| Single thread | Multiple threads |
| I/O bound     | CPU bound        |

‚ùå Interview Trap

    Is async faster?

    ‚úÖ No, it‚Äôs more efficient for waiting tasks.

```python
import time

def task():
    time.sleep(2)
    print("Task done")

task()
print("Finished")
```

```python
import asyncio

async def task():
    await asyncio.sleep(2)
    print("Task done")

async def main():
    print("Start")
    await task()
    print("End")

# asyncio.run(main())

await main()

```

```python
import asyncio

async def task(name):
    await asyncio.sleep(2)
    print(name, "done")

async def main():
    await asyncio.gather(
        task("A"),
        task("B")
    )

# asyncio.run(main()) ## vscode 
await main()
```

```python
import asyncio

async def task(name, delay):
    await asyncio.sleep(delay)
    print(name, "done")

async def main():
    await asyncio.gather(
        task("A", 2),
        task("B", 1),
    )

await main()

# üìå Notice:

# B finishes first

# Total time ‚âà 2 seconds
```

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print("Time:", end - start)
        return result
    return wrapper

@timer
def slow_func():
    time.sleep(2)

slow_func()
```

```python
import time
import asyncio

def async_timer(func):
    async def wrapper(*args, **kwargs):
        start = time.time()
        result = await func(*args, **kwargs)
        end = time.time()
        print("Time:", end - start)
        return result
    return wrapper

@async_timer
async def async_task():
    await asyncio.sleep(2)

await async_task()
```
