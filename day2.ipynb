# Python Tuples (Basic ‚Üí Advanced)

`Tuple is immutable and hashable, so it is safer and faster and can be used as dictionary keys.`

```python
t1 = (1, 2, 3)
t2 = 1, 2, 3          # parentheses optional
t3 = ()               # empty tuple
t4 = (5,)             # single element tuple (comma is MUST)
t = (5)               # int, NOT tuple
t = (5,)              # tuple ‚úÖ
```

`Tuple values cannot be changed after creation.`

Tuple is Immutable



```python
t = (1, 2, 3)
t[0] = 100
```

‚úÖ Tuple immutable

`‚ùå But mutable object inside tuple CAN change`

```python
t = (1, [2, 3])

t[1][0] = 100
print(t)   # (1, [100, 3])
```

`Tuple Methods (Only 2 üòÆ)`

üìå Interview:
Why only 2 methods?

‚û° Because tuple is immutable.

```python
t = (1, 2, 2, 3)

t.count(2)   # 2
t.index(3)   # 3
```

```python
t = (10, 20, 30)

for i in t:
    print(i)

for i in range(len(t)):
    print(t[i])
```

Tuple Packing & Unpacking

```python
t = 10, 20, 30  ## Packing

a, b, c = t     ## Unpacking
print(a, b, c)
```

```python
t = (1, 2, 3, 4, 5)  ## Extended Unpacking
a, *b, c = t
print(a)  # 1
print(b)  # [2, 3, 4]
print(c)  # 5
```

üî• INTERVIEW QUESTIONS (With Answers)

`Q1. Why tuples are faster than lists?`

Answer: Because tuples are immutable and have fixed memory.

`Q2. Can we modify a tuple?`

Answer: No, but mutable elements inside it can be modified.

`Q3. Why use tuple instead of list?`

Answer: Data safety, faster access, dictionary keys.

`Q4. Difference between (1) and (1,)?`

Answer: (1) ‚Üí int, (1,) ‚Üí tuple.

`Q5. Is tuple hashable?`

Answer: Yes, if all elements inside are hashable.

`Q6. How to swap variables using tuple?`

a, b = b, a

`Q7. Can tuple contain duplicate values?`

Answer: Yes.

`Q8. Can tuple contain different data types?`

t = (1, "hello", 3.5)

`Q9: Can tuple contain mutable objects?`

Answer: Yes, but tuple itself cannot be modified.

```python
## 1Ô∏è‚É£ Reverse a tuple

t = (1,32,4,5,7,4,0,8,9,5,1)
t = list(t)

t = t[::-1]
t = tuple(t)
rev = tuple(reversed(t))

# print('Reverse a tuple',t)

## 2Ô∏è‚É£ Find max & min from tuple

t = (1,32,4,5,7,4,0,8,9,5,1)
max_ = max(t)
min_ = min(t)
# print(max_)
# print(min_)


## 3Ô∏è‚É£ Remove duplicates from tuple

t = (1,32,4,5,7,4,0,8,9,5,1)
seen = set()
for i in t:
    if (t.count(i)>0):
        seen.add(i)
# print(seen)

t = list(t)
seen = []
for i in t:
    if i not in seen:
        seen.append(i)
# print(seen)
t = (1, 2, 2, 3, 1)  ## Remove Duplicates (Order matters!)
res = tuple(dict.fromkeys(t))
print(res)


# 4Ô∏è‚É£ Count element frequency
t = (1,32,4,5,7,4,0,8,9,5,1,1,1,1)
freq = {}
for i in t:
    #print(t.count(i))
    freq[i] = freq.get(i,0)+1  
print(freq)

# 5Ô∏è‚É£ Convert list of tuples to dictionary

t = (1,32,4,5,7,4,0,8,9,[5,1,1,1,1])
#t = list(t)
seen = {}
for i in t:
    if isinstance(i, list):
        i = tuple(i)  # Convert list to tuple to make it hashable
    seen[i] = seen.get(i, 0) + 1  # Increment the count

print(seen)

# 6Ô∏è‚É£ Sort tuple by second element

# Assuming t is a list of tuples, e.g., [(a, b), (c, d), ...]
# If t is flat, we can pair them up.

t = [(1,32), (4,5), (7,4), (0,8), (9,5), (1,1)]  # Example list of tuples

# Sort by the second element (index 1)
sorted_t = sorted(t, key=lambda x: x[1])
sorted_t = tuple(sorted(t, key=lambda x: x[1]))

print(sorted_t)

# 7Ô∏è‚É£ Check if tuple is palindrome

tv = (1,2,1)

if tv[::-1] == tv:
    print("palindrome")
else:
    print("not palindrome")

# Another method using loop
def is_palindrome(t):
    # ‡¶∂‡ßÅ‡¶ß‡ßÅ ‡¶Ö‡¶∞‡ßç‡¶ß‡ßá‡¶ï ‡¶™‡¶∞‡ßç‡¶Ø‡¶®‡ßç‡¶§ loop
    for i in range(len(t) // 2): 
        # ‡¶¨‡¶æ‡¶Æ ‡¶¶‡¶ø‡¶ï ‡¶ì ‡¶°‡¶æ‡¶® ‡¶¶‡¶ø‡¶ï‡ßá‡¶∞ element ‡¶§‡ßÅ‡¶≤‡¶®‡¶æ
        if t[i] != t[-i - 1]:
            return False
    # ‡¶∏‡¶¨ match ‡¶π‡¶≤‡ßá
    return True

print(is_palindrome(tv))  # True

## Can you check palindrome without indexing?

from collections import deque

def is_palindrome(t):
    d = deque(t)
    while len(d) > 1:
        if d.popleft() != d.pop():
            return False
    return True

```
