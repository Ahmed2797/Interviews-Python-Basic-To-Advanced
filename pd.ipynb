### 1Ô∏è‚É£2Ô∏è‚É£ Merge, Join, Concat 

Difference between merge and concat?

 - merge ‚Üí key based

 - concat ‚Üí axis based

```python
df1 = pd.DataFrame({
    "name": ["A", "B", "C"],
    "age": [20, np.nan, 21],
    "salary": [1000, 1500, 1200]
})

df2 = pd.DataFrame({
    "name": ["D", "E", "F"],
    "age": [22, 24, 23],
    "salary": [900, 1600, 1200]
})


pd.concat([df1, df2],axis=0)
pd.concat([df1, df2],axis=1)
# concat ‚Üí axis based
```

```python
df1 = pd.DataFrame({
    "name": ["A", "B", "C"],
    "age": [20, np.nan, 21],
    "salary": [1000, 1500, 1200]
})

df2 = pd.DataFrame({
    "name": ["A", "B", "C"],
    "age": [22, 24, 23],
    "salary": [900, 1600, 1200]
})

pd.merge(df1, df2, on="name", how="inner")
# merge ‚Üí key based
```

| how   | meaning   |
| ----- | --------- |
| inner | common    |
| left  | left all  |
| right | right all |
| outer | all       |

```python
import pandas as pd

data = {
    "emp_id": [101, 102, 103, 104, 105, 106],
    "name": ["A", "B", "C", "D", "E", "F"],
    "dept": ["IT", "HR", "IT", "HR", "IT", "Finance"],
    "salary": [5000, 4000, 6000, None, 5500, 4500],
    "experience": [2, 3, 5, 1, None, 4]
}

data = pd.DataFrame(data)
data
```

#### ‚ùì Find average salary per department (ignore nulls)

```python
## ‚ùì Find average salary per department (ignore nulls)
data.groupby('dept')['salary'].mean()
#data.groupby(['dept']).value_counts()
```

#### ‚ùì Find employees with salary greater than department average

 - transform() returns same-size Series ‚Üí perfect for filtering.

```python
display(data)

dept_avg = data.groupby("dept")["salary"].transform("mean")
data[data["salary"] > dept_avg]
```

```python
dp_avg = data.groupby('dept')['salary'].mean()
display(dp_avg) ## every department salary mean()
dept_avg = data.groupby("dept")["salary"].transform("mean")
display(dept_avg) 
## transform("mean") if in null value and fill every row its avg selary
```

#### ‚ùì Replace missing salary with department average

Why not use apply here?
