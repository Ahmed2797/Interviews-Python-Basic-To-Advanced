## 1. Measure of Central Tendency

Measures of central tendency describe the central or typical value of a dataset.

### 1.1 Mean
**Formula:**  
Mean = (Sum of all values) / (Number of values)

**Example:**  
Data: 2, 4, 6, 8  
Mean = (2 + 4 + 6 + 8) / 4 = **5**

---

### 1.2 Median
The middle value after sorting the data.

**Odd number of observations:**  
Data: 3, 5, 7  
Median = **5**

**Even number of observations:**  
Data: 2, 4, 6, 8  
Median = (4 + 6) / 2 = **5**

---

### 1.3 Mode
The value that appears most frequently.

**Example:**  
Data: 1, 2, 2, 3  
Mode = **2**

# üìä Statistics Notes

## 1. Measure of Central Tendency

Measures of central tendency describe the central or typical value of a dataset.

---

### 1.1 Mean

**Formula:**

$$
\text{Mean} = \frac{\sum x_i}{n}
$$

Where:  
- \( x_i \) = individual observations  
- \( n \) = number of observations  

**Example:**

Data: \( 2, 4, 6, 8 \)

$$
\text{Mean} = \frac{2 + 4 + 6 + 8}{4} = \frac{20}{4} = 5
$$

---

### 1.2 Median

The middle value after sorting the data.

**Odd number of observations:**

Data: \( 3, 5, 7 \)

$$
\text{Median} = 5
$$

**Even number of observations:**

Data: \( 2, 4, 6, 8 \)

$$
\text{Median} = \frac{4 + 6}{2} = 5
$$

---

### 1.3 Mode

The value that occurs most frequently.

**Example:**

Data: \( 1, 2, 2, 3 \)

$$
\text{Mode} = 2
$$

---

## 2. Measure of Dispersion

Measures of dispersion describe how spread out the data is.

---

### 2.1 Range

**Formula:**

$$
\text{Range} = \max(x) - \min(x)
$$

**Example:**

Data: \( 2, 4, 6, 10 \)

$$
\text{Range} = 10 - 2 = 8
$$

---

### 2.2 Variance

Variance measures the average squared deviation from the mean.

#### Population Variance

**Formula:**

$$
\sigma^2 = \frac{\sum (x_i - \mu)^2}{n}
$$

Where:  
- \( \mu \) = population mean  
- \( n \) = number of observations  

**Example:**

Data: \( 2, 4, 6 \)

$$
\mu = \frac{2 + 4 + 6}{3} = 4
$$

$$
\sigma^2 = \frac{(2-4)^2 + (4-4)^2 + (6-4)^2}{3}
$$

$$
\sigma^2 = \frac{4 + 0 + 4}{3} = 2.67
$$

---

### 2.3 Standard Deviation

Standard deviation is the square root of variance.

**Formula:**

$$
\sigma = \sqrt{\sigma^2}
$$

**Example:**

$$
\sigma = \sqrt{2.67} \approx 1.63
$$

## 3. Percentile and Quantile

---

## 3.1 Percentile

A percentile represents the value below which a given percentage of observations fall.

**k-th Percentile Formula (Position):**

$$
P_k = \frac{k}{100} \times (n + 1)
$$

Where:  
- \(k\) = percentile (10, 25, 70, etc.)  
- \(n\) = number of data points  

---

## Example 1: 10th Percentile

Data: \(5, 7, 12, 15, 18, 20, 22, 25\)  

Number of data points: \(n = 8\)  

**Step 1: Find position**

$$
P_{10} = \frac{10}{100} \times (8 + 1) = 0.1 \times 9 = 0.9
$$

**Step 2: Interpolate between 1st and 2nd value**

$$
P_{10} = X_1 + 0.9 \times (X_2 - X_1)
$$

$$
P_{10} = 5 + 0.9 \times (7 - 5) = 5 + 1.8 = 6.8
$$

‚úÖ **10th Percentile = 6.8**

---

### Example 2: 25th Percentile (Q1)

Data: \(5, 7, 12, 15, 18, 20, 22, 25\)  

Number of data points: \(n = 8\)  

**Step 1: Find position**

$$
P_{25} = \frac{25}{100} \times (8 + 1) = 0.25 \times 9 = 2.25
$$

**Step 2: Interpolate between 2nd and 3rd value**

$$
P_{25} = X_2 + 0.25 \times (X_3 - X_2)
$$

$$
P_{25} = 7 + 0.25 \times (12 - 7) = 7 + 1.25 = 8.25
$$

‚úÖ **25th Percentile = 8.25**

---

### 3.2 Percentile Rank

Percentile Rank tells **the percentage of data values below a given value** \(X\).

**Formula:**

$$
\text{Percentile Rank of X} = \frac{\text{Number of values below X}}{n} \times 100
$$

Where:  
- \(n\) = total number of observations  

---

## Example: Percentile Rank of 10

Data: \(5, 7, 12, 15, 18, 20, 22, 25\)  

- Count values below 10: \(5, 7\) ‚Üí 2 values  
- Total \(n = 8\)  

$$
\text{Percentile Rank of 10} = \frac{2}{8} \times 100 = 25\%
$$

‚úÖ **Percentile Rank of 10 = 25th percentile**

> **Observation:**  
> Percentile value ‚Üí value corresponding to a percentile  
> Percentile rank ‚Üí percentile corresponding to a value
