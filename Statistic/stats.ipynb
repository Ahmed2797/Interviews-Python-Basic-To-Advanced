## 4Ô∏è‚É£ Mean, Median, Mode

* **Mean** ‚Üí Average

## 1Ô∏è‚É£ What is Statistics and Data?

### What is Data?

Data is a collection of **raw facts, values, or observations** that can be numbers, text, images, or measurements.

**Examples:**

* Exam scores of students
* Heights and weights of people
* Daily temperature readings

---

### What is Statistics?

Statistics is the **science of collecting, organizing, analyzing, and interpreting data** to make decisions or conclusions.

Statistics helps us:

* Understand large datasets
* Identify patterns and trends
* Make predictions and decisions

---

## 2Ô∏è‚É£ Types of Statistics

There are **two main types of statistics**:

1. **Descriptive Statistics**
2. **Inferential Statistics**

---

### 2.1 Descriptive Statistics

Descriptive Statistics is used to **summarize and describe the main features of a dataset**. It does not make predictions beyond the given data.

**Common tools:**

* Mean, Median, Mode
* Variance, Standard Deviation
* Tables, Charts, Graphs

**Example:**

> The average salary of 10 employees is $40,000.

---

### 2.2 Inferential Statistics

Inferential Statistics is used to **draw conclusions or make predictions about a population** based on a sample.

**Common tools:**

* Hypothesis Testing
* Confidence Intervals
* Z-test, T-test
* P-value

**Example:**

> From 100 voters, predicting the election result of an entire country.

---

## 3Ô∏è‚É£ Population and Sample

### Population

A **population** is the complete set of all individuals or items that we want to study.

**Example:**

> All students in a university.

---

### Sample

A **sample** is a smaller subset taken from the population to represent it.

**Example:**

> 200 students selected from a university of 10,000 students.

üìå Statistics and Machine Learning mostly work with **samples**, not full populations.

---

## 4Ô∏è‚É£ Types of Sampling (With Examples)

### 1. Simple Random Sampling

Every member has an **equal chance** of being selected.

**Example:**

> Randomly selecting 50 students from a class list.

---

### 2. Systematic Sampling

Selecting items at a **fixed interval**.

**Example:**

> Choosing every 10th customer entering a store.

---

### 3. Stratified Sampling

Population is divided into **groups (strata)**, and samples are taken from each group.

**Example:**

> Selecting students from each department proportionally.

---

### 4. Cluster Sampling

Population is divided into clusters, and **entire clusters** are selected.

**Example:**

> Selecting 5 schools randomly and surveying all students in those schools.

---

## 5Ô∏è‚É£ Variance

### What is Variance?

Variance measures **how far data values are spread out from the mean**.

* Low variance ‚Üí data points are close to the mean
* High variance ‚Üí data points are widely spread

---

### Types of Variance

#### 1. Population Variance (œÉ¬≤)

Used when data includes the **entire population**.

**Formula:**
œÉ¬≤ = Œ£(x ‚àí Œº)¬≤ / N

**Example:**

> Calculating variance of marks of all students in a class.

---

#### 2. Sample Variance (s¬≤)

Used when data is a **sample from a population**.

**Formula:**
s¬≤ = Œ£(x ‚àí xÃÑ)¬≤ / (n ‚àí 1)

**Example:**

> Calculating variance of marks of 50 selected students.

üìå Sample variance uses **(n ‚àí 1)** to reduce bias.

---

## üéØ Why These Concepts Matter (ML & Data Science)

* Sampling affects model accuracy
* Variance helps detect overfitting and underfitting
* Inferential statistics validates model results

## 6Ô∏è‚É£ What is a Variable?

A **variable** is any characteristic, feature, or quantity that can **change or vary** from one observation to another.

In statistics and data science, variables store **information about data** that we want to analyze.

**Examples:**

* Age of students
* Height of people
* Marks in an exam
* Gender of customers

---

## 7Ô∏è‚É£ Types of Variables

Variables are mainly classified into **two broad categories**:

1. **Qualitative (Categorical) Variables**
2. **Quantitative (Numerical) Variables**

---

## 7.1 Qualitative (Categorical) Variables

Qualitative variables describe **qualities or categories**, not numbers.

### a) Nominal Variables

* Categories with **no natural order**

**Examples:**

* Gender (Male, Female)
* Blood group (A, B, AB, O)
* Color (Red, Blue, Green)

üìå Used in classification problems.

---

### b) Ordinal Variables

* Categories with a **meaningful order**, but differences are not measurable

**Examples:**

* Education level (School < College < University)
* Customer satisfaction (Low, Medium, High)
* Movie ratings (‚≠ê to ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê)

---

## 7.2 Quantitative (Numerical) Variables

Quantitative variables represent **numerical values** and allow mathematical operations.

---

### a) Discrete Variables

* Countable numbers

**Examples:**

* Number of students in a class
* Number of cars in a parking lot
* Number of emails received

---

### b) Continuous Variables

* Can take **any value within a range**

**Examples:**

* Height of a person
* Weight of an object
* Temperature of a city

---
