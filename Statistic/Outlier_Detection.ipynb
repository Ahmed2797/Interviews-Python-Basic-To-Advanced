# ðŸ“Š Outlier Detection â€“ Formula Sheet

This document contains **all major outlier detection formulas** used in **Data Science & Machine Learning**.

---

## 1ï¸âƒ£ IQR Method (Interquartile Range)

### Formulas

$$
Q_1 = 25^{th}\ percentile
$$

$$
Q_3 = 75^{th}\ percentile
$$

$$
IQR = Q_3 - Q_1
$$

$$
Lower\ Bound = Q_1 - k \times IQR
$$

$$
Upper\ Bound = Q_3 + k \times IQR
$$

**Typical value:**

$$
k = 1.5
$$

### Outlier Condition

$$
x < Lower\ Bound \quad \text{or} \quad x > Upper\ Bound
$$

**Best for:** Skewed data (salary, income)

---

## 2ï¸âƒ£ Mean Â± Standard Deviation Method

### Formulas

$$
\mu = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

$$
\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2}
$$

$$
Lower\ Bound = \mu - k\sigma
$$

$$
Upper\ Bound = \mu + k\sigma
$$

**Common values:**

$$
k = 2 \quad \text{or} \quad k = 3
$$

### Outlier Condition

$$
x < \mu - k\sigma \quad \text{or} \quad x > \mu + k\sigma
$$

**Best for:** Normally distributed data

---

## 3ï¸âƒ£ Z-Score Method

### Formula

$$
Z = \frac{x - \mu}{\sigma}
$$

### Outlier Condition

$$
|Z| > k
$$

**Typical threshold:**

$$
k = 3
$$

**Meaning:** Measures how many standard deviations a value is from the mean.

---

## 4ï¸âƒ£ Quantile (Percentile) Method

### Formulas

$$
Lower\ Bound = Q_p
$$

$$
Upper\ Bound = Q_{1-p}
$$

**Common choice:**

$$
p = 0.05 \Rightarrow (5\%, 95\%)
$$

### Outlier Condition

$$
x < Q_{0.05} \quad \text{or} \quad x > Q_{0.95}
$$

**Best for:** Large datasets, fast filtering

---

## 5ï¸âƒ£ Capping / Winsorization

### Formula

$$
x' =
\begin{cases}
Lower\ Bound, & x < Lower\ Bound \\
x, & Lower\ Bound \le x \le Upper\ Bound \\
Upper\ Bound, & x > Upper\ Bound
\end{cases}
$$

**Purpose:** Limits extreme values instead of removing them.

---

## 6ï¸âƒ£ Modified Z-Score (Robust Method)

### Formulas

$$
\tilde{x} = \text{Median}
$$

$$
MAD = \text{median}(|x_i - \tilde{x}|)
$$

$$
Z_{mod} = \frac{0.6745(x - \tilde{x})}{MAD}
$$

### Outlier Condition

$$
|Z_{mod}| > 3.5
$$

**Best for:** Data with extreme skew or outliers

---

## ðŸŽ¯ Interview Tip

> **Use IQR for skewed data, Z-score for normal data, and capping when data loss is not acceptable.**


```python
import numpy as np
import pandas as pd

def outlier_detection(
    df: pd.DataFrame,
    method: str = "iqr",
    threshold: float = 1.5,
    columns: list = None,
    lower_q: float = 0.05,
    upper_q: float = 0.95
):
    """
    Outlier detection and handling function
    """

    df = df.copy()

    # Select numeric columns only
    if columns is None:
        columns = df.select_dtypes(include=np.number).columns

    for col in columns:
        col_data = df[col]

        if method == "iqr":
            q1 = col_data.quantile(0.25)
            q3 = col_data.quantile(0.75)
            iqr = q3 - q1

            lower = q1 - threshold * iqr
            upper = q3 + threshold * iqr

            df = df[(col_data >= lower) & (col_data <= upper)]

        elif method == "std":
            mean_val = col_data.mean()
            std_val = col_data.std()

            lower = mean_val - threshold * std_val
            upper = mean_val + threshold * std_val

            df = df[(col_data >= lower) & (col_data <= upper)]

        elif method == "zscore":
            std_val = col_data.std()
            if std_val == 0:
                continue

            z_score = np.abs((col_data - col_data.mean()) / std_val)
            df = df[z_score <= threshold]

        elif method == "quantile":
            lower = col_data.quantile(lower_q)
            upper = col_data.quantile(upper_q)

            df = df[(col_data >= lower) & (col_data <= upper)]

        elif method == "capping":
            q1 = col_data.quantile(0.25)
            q3 = col_data.quantile(0.75)
            iqr = q3 - q1

            lower = q1 - threshold * iqr
            upper = q3 + threshold * iqr

            df[col] = np.clip(col_data, lower, upper)

        else:
            raise ValueError("Invalid method")

    return df
```
