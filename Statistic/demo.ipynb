    üß† ‡¶∏‡¶π‡¶ú ‡¶ï‡¶•‡¶æ:

```python
import matplotlib.pyplot as plt

# generate power law data
data = np.random.pareto(a=2.5, size=10000) + 1

plt.hist(data, bins=100)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

plt.xscale('log')
plt.yscale('log')
plt.hist(data, bins=100)
plt.show()
```

### Q-Q plot

```python
#### Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(8,4))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
s=np.random.normal(0.5,0.2,1000) ## mean,std

plot_data(s)
```

```python
s=np.random.lognormal(0.5,0.2,1000) ## mean,std

plot_data(s)
```
