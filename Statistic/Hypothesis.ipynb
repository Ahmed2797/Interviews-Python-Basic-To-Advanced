## 4. Hypothesis Testing: Z-Test and T-Test

Hypothesis testing is used to determine whether a sample data provides enough evidence to infer something about a population.

---

### **4.1 Z-Test**

Used when:  
- Population standard deviation (\(\sigma\)) is **known**  
- Sample size \(n \ge 30\)  

**Formula for One-Sample Z-Test:**

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$

Where:  
- \(\bar{X}\) = sample mean  
- \(\mu\) = population mean  
- \(\sigma\) = population standard deviation  
- \(n\) = sample size  

**Example:**

- Population mean (\(\mu\)) = 50  
- Population std (\(\sigma\)) = 10  
- Sample mean (\(\bar{X}\)) = 54  
- Sample size (\(n\)) = 36  

$$
Z = \frac{54 - 50}{10 / \sqrt{36}} = \frac{4}{10/6} = \frac{4}{1.6667} \approx 2.4
$$

Interpretation: Compare \(Z\) to critical Z-value from table (e.g., 1.96 for 5% significance).

---

### **4.2 T-Test**

Used when:  
- Population standard deviation (\(\sigma\)) is **unknown**  
- Sample size \(n < 30\)  

**Formula for One-Sample T-Test:**

$$
t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
$$

Where:  
- \(s\) = sample standard deviation  
- Other symbols same as Z-test  

**Example:**

- Population mean (\(\mu\)) = 50  
- Sample mean (\(\bar{X}\)) = 54  
- Sample std (\(s\)) = 12  
- Sample size (\(n\)) = 16  

$$
t = \frac{54 - 50}{12 / \sqrt{16}} = \frac{4}{12/4} = \frac{4}{3} \approx 1.33
$$

Interpretation: Compare \(t\) with critical t-value from t-table (df = n-1 = 15).

---

### **Summary Table**

| Test      | σ known? | Sample size | Formula                                   |
|-----------|-----------|------------|------------------------------------------|
| Z-Test    | Yes       | ≥ 30       | \( Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} \) |
| T-Test    | No        | < 30       | \( t = \frac{\bar{X} - \mu}{s / \sqrt{n}} \)      |

---

**Tip:**  

- Use **Z-Test** for large samples or known σ  
- Use **T-Test** for small samples or unknown σ  
- Always compare test statistic with **critical value** or calculate **p-value**
