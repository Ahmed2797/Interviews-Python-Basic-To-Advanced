#### Matrix Operations

ðŸ‘‰ Transpose :

- row â†’ column

- column â†’ row

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print(A @ B) # matrix multiplication
print(np.dot(A,B)) # same
print(np.transpose(A)) # transpose
print(np.linalg.inv(A)) # inverse
```

`np.transpose(A) swaps rows and columns of a matrix. It is used extensively in linear algebra and machine learning for matrix multiplication and data transformation.`

```python
print(np.transpose(A)) # transpose
A
```

#### Aggregation / Statistics

```python
a = np.array([1,2,3,4,5])
print(np.sum(a))
print(np.mean(a))
print(np.median(a))
print(np.std(a))
print(np.var(a))
print(np.max(a))
print(np.min(a))
```

#### ML-focused
##### Normalize a dataset (Z-score)

```python
arr = np.array([1,2,3,4,5])
normalized = (arr - np.mean(arr)) / np.std(arr)
print(normalized)
```



##### Standardize features



```python
X = np.random.rand(5,3)
X_std = (X - X.mean(axis=0)) / X.std(axis=0)
print(X_std)
```

##### Linear regression formula

    Î²=(X.T X)âˆ’1 X.T y 

    

```python
# y = Xb
X = np.random.rand(5,3)
y = np.random.rand(5,1)
b = np.linalg.inv(X.T @ X) @ X.T @ y
print(b)
```

##### Vectorized sigmoid

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z = np.array([-1,0,1])
print(sigmoid(z))
```

#### Eigenvalues & Eigenvectors.

```python
A = np.array([[1,2],[2,3]])
vals, vecs = np.linalg.eig(A)
print(vals)
print(vecs)
```
